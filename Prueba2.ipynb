```python
vendedores = ['Ana', 'Luis', 'Marta', 'Sergio']
ventas = [1500.50, 2300.75, 1800.00, 2200.25]
Resumen = list(zip(vendedores, ventas))
for vendedor, venta in Resumen:
    print(f'{vendedor}: ${venta:,.2f}')
```
